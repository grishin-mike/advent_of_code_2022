## Part 1:

```
import numpy as np
```

```
lines = open('Data/data_day_09.txt').read().split('\n')
```

```
def calculate_tail():
    
    global head, tail, path
    
    diff_vect = head - tail
    
    if diff_vect[0]==2:
        tail = tail + [1,0]
        tail[1] = head[1]
    
    if diff_vect[1]==2:
        tail = tail + [0,1]
        tail[0] = head[0]
    
    if diff_vect[0]==(-2):
        tail = tail + [-1,0]
        tail[1] = head[1]
        
    if diff_vect[1]==(-2):
        tail = tail + [0,-1]
        tail[0] = head[0]
     
    path.add(tuple(tail))
    
def move_head(path_vect, num):
    
    global head,tail,path
    
    if path_vect=='R':
        vect = np.array([0,1])
    elif path_vect=='L':
        vect = np.array([0,-1])
    elif path_vect=='U':
        vect = np.array([1,0])
    elif path_vect=='D':
        vect = np.array([-1,0])
    
    for i in range(num):
        head = head+vect
        calculate_tail()
        
```

```
head = np.array([0,0])
tail = np.array([0,0])

path = set()

for line in lines:
    lst = line.split()
    move_head(lst[0], int(lst[1]))

print(f'Answer par 1: {len(path)}')
```
