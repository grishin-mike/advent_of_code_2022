# Day 2: Rucksack Reorganization (solution)
Task conditions: https://adventofcode.com/2022/day/3

## Part one

```
#  create a dictionary with sums of letters 
import string

dict = {}
i=1

for _ in string.ascii_lowercase:
    dict[_]=i
    i+=1

for _ in string.ascii_uppercase:
    dict[_]=i
    i+=1
    
print(dict)
```

```
# returns sum of the string
def find_elem(first_part, second_part):
     
    for i in first_part:
        for j in second_part:
            if i==j:
                return(dict[j])
```

```
sum = 0

with open("Data/data_day_03.txt", "r") as file1:
    
    for line in file1:
        ln = line.strip()
        n = len(ln) // 2
        first_part = ln[:n]
        second_part = ln[n:]
        
        sum += find_elem(first_part, second_part)

print(f"First part answer: {sum}")
```

# Second part

```
def find_sum(lst):
    
    for i in lst[0]:
        for j in lst[1]:
            if j==i:
                for k in lst[2]:
                    if k==j:
                        return dict[k]
```

```
with open("Data/data_day_03.txt", "r") as file1:
    
    lst = []
    iter= 1
    sum = 0
    
    for line in file1:
        if iter<3:
        
            lst.append(line.strip())
            iter+=1
        
        elif iter==3:
            
            lst.append(line.strip())
            sum = sum+find_sum(lst)
            
            iter=1
            lst=[]
            
print(f"Second part answer: {sum}")
```
