Task conditions: https://adventofcode.com/2022/day/4


```
def res(first_part,second_part):
    if (int(first_part[0])<= int(second_part[0])) and (int(first_part[1])>= int(second_part[1])):
        return 1
    if (int(first_part[0])>= int(second_part[0])) and (int(first_part[1])<= int(second_part[1])):
        return 1
    return 0

def res_2(first_part,second_part):
    #1
    if (int(first_part[0])<= int(second_part[0])) and (int(first_part[1])>= int(second_part[1])):
        return 1
    #2
    if (int(first_part[0])>= int(second_part[0])) and (int(first_part[1])<= int(second_part[1])):
        return 1
    
    if (int(first_part[0])>= int(second_part[0])) and (int(first_part[0])<= int(second_part[1])):
        return 1
    if (int(first_part[1])>= int(second_part[0])) and (int(first_part[1])<= int(second_part[1])):
        return 1
    
    return 0
```

```
import re

sum=0

with open("Data/data_day_04.txt", "r") as file1:
    
    for line in file1:
        ls=line.strip()
        comma = ls.find(',')
        first_part = ls[:comma]
        second_part = ls[comma+1:]
        first_part = re.split('-',first_part)
        second_part = re.split('-',second_part)
        
        sum += res(first_part,second_part)
    
print(f'First part answer: {sum}')  
```
