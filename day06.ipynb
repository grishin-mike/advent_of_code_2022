## Part 1&2

```
def check_list(lst, num_elem):
    
    if len(set(lst))==num_elem: # python set contains only unique elements
        return True
    return False 

def find_sequance(file_name, num_elem):
    
    f = open(file_name,'r')
    
    num_char = 1
    lst = []

    break_cicle = False

    for i in f.read():
   
        if num_char < num_elem:
        
            lst.append(num_elem)
    
        elif num_char == num_elem:
        
            lst.append(i)
            break_cicle = check_list(lst, num_elem)
    
        else:
        
            lst.pop(0)
            lst.append(i)
            break_cicle = check_list(lst ,num_elem)
    
        if break_cicle:
            
            print(f'List {lst} num: {num_char}')
            break
    
        num_char+=1      
```

```
file_name = 'Data/data_day_06.txt'
find_sequance(file_name, 4) # answer to part one
find_sequance(file_name, 14)# answer to part one
```
