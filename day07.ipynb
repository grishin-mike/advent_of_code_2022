# Day 7: No Space Left On Device

Task conditions: https://adventofcode.com/2022/day/7

## Part 1:

```
def sum_cat_f(sum_catalog, cat, sum_cat):
    
    cat1 = cat.copy()
    
    is_sum = False
    
    for elem in sum_catalog:
        if elem[0]==cat1:
            elem[1]+=sum_cat
            is_sum = True
    
    if is_sum == False:
        sum_catalog.append([cat1, sum_cat])

def sum_size(sum_catalog, current_catalog, sum_cat):
        
    cat = current_catalog.copy()
    
    while cat!=[]:
        sum_cat_f(sum_catalog, cat, sum_cat)
        cat.pop()   
```

```
lines = open('Data/data_day_07.txt').read().split('\n')

current_catalog=[] # текущий каталог
sum_catalog = [] # сумма по каждому каталогу ([[], ])


for line in lines:
    elem = line.split()
    
    if elem[0]=='$':
        #print(elem)
        if elem[1] == 'cd':
            if elem[2] == '..':
                current_catalog.pop() 
            else:
                current_catalog.append(elem[2])
    else:
        if elem[0]!='dir':
            sum_size(sum_catalog, current_catalog, int(elem[0]))
```

```
sum = 0
for i in sum_catalog:
    if i[1]<=100000:
        sum+=i[1]
        
print(f'Answer part 1: {sum}')
```

## Part 2:

```
total_space = 70000000

free_space = total_space - sum_catalog[0][1]

remaning_space = 30000000-free_space

#print(remaning_space)

min = 30000000

for _ in sum_catalog:
    if _[1] >= remaning_space:
        if _[1] < min:
            min=_[1]
print('Answer part 2: {}'.format(min))
```
