# Day 10: Cathode-Ray Tube
Task conditions: https://adventofcode.com/2022/day/10

## Part 1:

```
lines = open('Data/data_day_10.txt').read().split('\n')

i = 1
lst = []

for line in lines:
    l = line.split()
    
    if l[0]=='noop':
        lst.append(i)
    elif l[0]=='addx':
        lst.append(i)
        i = i + int(l[1])
        lst.append(i)
  
```

```
list_ten = [20,60,100,140,180,220]
```

```
sum = 0

for i in list_ten:
    sum=sum+i*lst[i-2]

print(f'Answer of part 1: {sum}')
```

# Part 2:
