# Day 5: Supply Stacks

Task conditions: https://adventofcode.com/2022/day/5

## Part 1:

```
import re

def init_cargo():
    return [['W', 'B', 'D', 'N', 'C', 'F', 'J'],
            ['P','Z','V','Q','L','S','T'],
            ['P','Z','B','G','J','T'],
            ['D','T','L','J','Z','B','H','C'],
            ['G','V','B','J','S'],
            ['P','S','Q'],
            ['B','V','D','F','L','M','P','N'],
            ['P','S','M','F','B','D','L','R'],
            ['V','D','T','R']
           ] 
```

```
# function for box moving
def transfer(i_1,i_2):
    elem = list_cargo[i_1].pop()
    list_cargo[i_2].append(elem)
```

```
list_cargo = init_cargo()

with open("Data/data_day_05.txt", "r") as file1:
    
    for line in file1:
        
        line_list = re.split(' ',line.strip())
        
        count = int(line_list[1])
        stack_1 = int(line_list[3])-1
        stack_2 = int(line_list[5])-1
        
        for i in range(count):
            transfer(stack_1,stack_2)
 
st = ""
for i in list_cargo:
    st = st + i[-1]

print(f'First part answer: {st}')
```

## Part 2:

```
# NEW function for box moving
def transfer_2(stack_1,stack_2, count):
    elems = list_cargo[stack_1][-(count):]
    list_cargo[stack_2].extend(elems)
    del list_cargo[stack_1][-(count):]
```

```
list_cargo = init_cargo()

with open("Data/data_day_05.txt", "r") as file1:
    
    for line in file1:
        
        line_list = re.split(' ',line.strip())
        
        count = int(line_list[1])
        stack_1 = int(line_list[3])-1
        stack_2 = int(line_list[5])-1
    
        #print(f"count {count}, {stack_1}, {stack_2}")
        
        transfer_2(stack_1,stack_2, count)

st = ""
for i in list_cargo:
    st = st + i[-1]

print(f'Second part answer: {st}')
```
