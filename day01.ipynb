## Part one & part two
The elves changed the terms of reference during development: at the beginning they wanted to get the largest amount of calories from the list, and then they wanted to get the three largest amounts. Now they should be happy, because using the **sum_elves** function, they can get the sum of any number of the largest elements.
 
I implemented the function to avoid software changes in case of changes in the requirements for the number of elves.


```
# add item in ordered list
def edit_list(ls, num):
    for i in range(len(ls)):
        if ls[i] < num:
            ls[i], num = num, ls[i]

             
# num_elves - how many elves to count
def sum_elves(filename, num_elves):
    
    file1 = open(filename, "r")

    ls = [0 for r in range(num_elves)] #list of best elves

    res = 0

    while True:
        
        line = file1.readline()
  
        if not line:
            edit_list(ls, res)
            break
    
        if len(line.strip()) == 0:
            edit_list(ls, res)
            res = 0
    
        else:
            res += int(line.strip())
    
    ls = sum(ls)
    
    print(f'Elves to count: {num_elves} sum = {ls}')

    file1.close   
```

```
#Part 1: one elve to count 
sum_elves("Data/data_day_01.txt",1)

#Part 2: three elves to count 
sum_elves("Data/data_day_01.txt",3)

#And so on
sum_elves("Data/data_day_01.txt",5)
```
