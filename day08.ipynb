# Day 7: Treetop Tree House
Task conditions: https://adventofcode.com/2022/day/8

## Part 1:

```
import numpy as np
```

```
lines = open('Data/data_day_08.txt').read().split('\n')
```

```
arr = []

for i in lines:
    arr.append(list(i))
    
trees_array = np.array(arr).astype(np.int)
```

```
length, width = trees_array.shape
```

```
def is_larger(tree, np_array):
    
    for elem in np_array:
        if elem>=tree:
            return True
    return False
    
```

```
def is_visible(tree, i, j):
    
    arr_left  = trees_array[i,:j]
    arr_right= trees_array[i,(j+1):]
    
    arr_up = trees_array[:i,j]
    arr_down = trees_array[(i+1):,j]
    
    a = is_larger(tree,arr_left)
    b = is_larger(tree,arr_right)
    c = is_larger(tree,arr_up)
    d = is_larger(tree,arr_down)
   
    if a*b*c*d==0:
        return True
    return False
```

```
sum = 0

for i in range(length):
    
    for j in range(width):
    
        tree = trees_array[i,j]
        sum+=is_visible(tree,i,j)

print(f'Answer of part 1: {sum}')           
```

## Part 2:

```
def num_trees(tree, np_array):
    count = 0
    for i in np_array:
        count+=1
        if i>=tree:
            break
    return count
```

```
def scenic_score(tree, i, j):
    
    arr_left  = trees_array[i,:j]
    arr_left  = arr_left[::-1] #
    arr_right = trees_array[i,(j+1):]
    
    arr_up = trees_array[:i,j]
    arr_up = arr_up[::-1] #
    arr_down = trees_array[(i+1):,j]
    
    a = num_trees(tree,arr_left)
    b = num_trees(tree,arr_right)
    c = num_trees(tree,arr_up)
    d = num_trees(tree,arr_down)
   
    return a*b*c*d   
```

```
max_sum = 0

for i in range(length):
    
    for j in range(width):
    
        tree = trees_array[i,j]
        score = scenic_score(tree,i,j)
        if score>max_sum:
            max_sum=score
        
print(f'Answer of part 2: {max_sum}') 
```
